7. 为找到的这些state/region的state项补上正确的值，从而去除掉state这一列的所有NaN！记住这样清除缺失数据NaN的方法！
     - pop_abbr['state/region']=='PR'
     - pop_abbr.loc[pop_abbr['state/region']=='PR','state'] = 'PUERTO RICO'

```python
def map_na(df,to_filedname,mapper_field,**mapper):
    """
    df 修改的DataFrame对象
    to_fieldname 修改nan值字段的名字
    mapper_field 查询有nan值行的映射字段
    mapper mapper_field字段的值与替换nan的数值映射的dict,
    如：
    abbr_pops.loc[abbr_pops['state/region'] == 'PR','state'] = 'PUERTO RICO'
    abb_pops 即是df
    state/region   mapper_field
    state   to_fieldname
    PUERTO RICO    是 mapper_field字段中PR值对应的to_fieldname的数值
    """
    for key,value in mapper.items():
        df.loc[df[mapper_field]==key,to_fieldname]=value
```
