```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame,MultiIndex
```

## Series的多层索引
- 隐式方式：直接使用多级的list指定
- 显示方式： 使用MultiIndex类的from_arrays()from_tuples()/from_product()方法创建


```python
#from_product()
s1 = Series(np.random.randint(1000,9000,size=6),
           index = MultiIndex.from_product([['西安','北京','上海'],
                                           ['vivo','oppo']]))
s1
```

DataFrame的多层索引
- 多层索引的创建同Series
- DataFrame可以创建多层列索引

```python
#题目二： 创建四个城市的五种手机的四个季度（3个月）的销售情况
#         城市：  杭州/重庆/大连/丽江
#         手机：   HuaWei P30,MI9,Vivo 27,Oppo FindX, Oneplus 7 pro
index = MultiIndex.from_product([['杭州','重庆','大连','丽江'],
                                 ['HuaWei P30','MI9','Vivo 27','Oppo FindX', 'Oneplus 7 pro']])
columns = MultiIndex.from_product([['一季度','二季度','三季度','四季度'],
                                   ['月1','月2','月3']])
data = np.random.randint(1000,9000,size=(20,12))
df2 = DataFrame(data,index,columns)
df2
```

```python
#多层索引的操作与切片
- 最外层的表示第一级层索引，level=0
- 最内层的表示最末级层次索引，level=-1 ，如果层次为2曾，也可以是-1
- 选择某一索引时，必须要写全(以元组方式)，如('西安','Oppo')
```

```python
# 查看大连的第一季度的销量情况
df2.loc['大连','一季度']
```

```python
#查看杭州的Vivo 27的‘一季度’和‘三季度’的销量情况
df2.loc[('杭州','Vivo 27'),['一季度','三季度']]
```

```python
#显示第三行、第四行、第九行的‘一季度’的前2月的销售情况
df2.iloc[[2,3,8],:2]
```

练习：
创建一个DataFrame，表示出张三李四期中期末各科成绩

```python
sdf1 = DataFrame(np.random.randint(150,size=(2,6)),
                columns=MultiIndex.from_product([['期中','期末'],
                                                ['语','数','外']]),
                index=['张三','李四'])
sdf1
```

```python
df2.stack()  #默认情况，将列的最内层的索引转成行的最内层的索引
```

```python
df2.stack(level=0) #将列的最外层索引转成行的最内层的索引
```

```python
df2.unstack()#将行的最内层索引转换成列的最内层的索引
```

```python
df2.unstack().sum(axis=1)
```

```python
#统计不同的手机的全年销量情况
df2.unstack(level=0).sum(axis=1)
```

#多层索引下的聚合操作
- min()/max()/mean()/sum()聚合函数中可以指定level，表示按哪一层次进行聚合

```python
#查看不同城市的不同手机的各个季度的销量情况
df2.sum(axis=0,level=0).sum(axis=1,level=0)
```

```python
sdf1.mean(axis=1,level=-1)
```
