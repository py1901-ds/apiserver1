```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame,MultiIndex
```

```python
df1 = DataFrame(np.random.randint(100,size=(5,3)),
               index=list('ABCDE'),
               columns=['Python','H5','Testing'])
df2 = DataFrame(np.random.randint(100,size=(5,3)),
               index=list('BCDEF'),
               columns=['Python','Java','Testing'])
display(df1,df2)
```

```python
#将df1和df2纵向级联，行数应该是5+5，列则会相对于df1增加新一个列Java
#将合并后的Nan数据按行后置填充，补全数据
pd.concat((df1,df2),axis=0,sort=True).fillna(method='bfill',axis=1)
```

```python
#将df1和df2横向级联,列数应该是3+3，行则会相对于df1，增加一个新的行F
pd.concat((df1,df2),axis=1,sort= True,keys=['期中','期末'],join_axes=[pd.Index(['B','D'])])
```

```python
##合并merge()
- 类似于数据库的join连接操作，即根据内容进行关联
- 存在1对1，1对多或多对1，多对多的连接关系
```

```python
emp1 = DataFrame({
    'employee':['Lisa','Bob','Jake'],
    'group':['accounting','engineering','hr']
})
emp2 = DataFrame({
    'employee':['Lisa','Bob','Jake'],
    'hire_date':['2010','2014','2018']
})
```

```python
emp1
```

```python
emp2
```

```python
pd.merge(emp1,emp2)#按员工的姓名进行合并，how='inner'内链接
```

```python
emp3 = DataFrame({
    'employee':['Disen','Bob','Jarrey'],
    'salary':['2000','1000','1800']
})
```

```python
emp3
```

```python
pd.merge(emp1,emp3)
```

```python
#将emp1和emp3进行合并，并且保留所有的原表数据
# how：{'left','right','outer','inner'},
pd.merge(emp1,emp3,how='left')
```

```python
emp4 = DataFrame({
    'group':['engineering','engineering','hr'],
    'supervisor':['Carly','Guibo','Steven']
})
```

```python
emp4
```

```python
pd.merge(emp1,emp4)
```

```python
stu1 = DataFrame({
    'id':['1001','1002','1003'],
    'name':['Disen','Vva','Bob'],
    'city':['西安','上海','广州']
})
stu2 = DataFrame({
    'id':['1001','1002','1003'],
    'name':['Disen','Vva','Bob'],
    'score':[200,300,230]
})
```

```python
stu1
```

```python
stu2
```

```python
#使用name字段关系两个stu1，stu2表
pd.merge(stu1,stu2,on='name',suffixes=['_left','_right'])
```

```python
city1 = DataFrame({
    'city_name':['西安','北京','成都','上海'],
    'salary':[9000,15000,9100,16000]
})
city2 = DataFrame({
    'city':['西安','北京','成都','上海'],
    'jobs':[900,3000,1200,2100]
})
```

```python
display(city1,city2)
```

```python
pd.merge(city1,city2,left_on='city_name',right_on='city')
```

```python
city3=pd.merge(city1,city2,left_index=True,right_index=True)
city3
```
