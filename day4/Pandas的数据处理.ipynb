### 删除重复元素
- drop()
- dropna() 删除存在NaN值的行或列
- drop_duplicates() 删除重复行
    - duplicated()  查看重复行

```python
# 创建项目管理中的项目表
df = DataFrame({
    'PID':['PXA1901-001','PXA1901-001','PXA1901-001'],
    'PNAME':['91理财APP','生态水果农产品服务平台','Bug管理平台'],
    'PMGR':['解迎接','世宏鹏','里亚京'],
    'PWEEK':[2,3,2],
    'PSTART':['2019-06-23','2019-06-21','2019-06-20'],
    'PINDEX':['http://91pp.com','http://fruit-bule.com','http://bug.com'],
    'PSOURCE':[np.nan,np.nan,np.nan],
    'PDOC':[np.nan,np.nan,np.nan]
})
df
```

```python
df2 = pd.concat((df,df),axis=0,ignore_index=True)
df2
```

```python
#删除重复的行
df2.drop_duplicates(keep='first',inplace=True)
df2
```

```python
df2 = pd.concat((df,df,df),axis=0,ignore_index=True)
df2
```

```python
df2.drop(df2[df2.duplicated(keep='last')].index)
```

```python
#删除指定的列
df.drop(columns=['PSOURCE','PDOC'])
```

```python
#删除所有值为NaN的列
df.dropna(axis=1,how='all')
```

```python
df.drop(columns=df.loc[:,df.isna().all(axis=0)].columns)
```

## 2.映射
- replace() 内容替换
- rename()  (index/columns)索引标签替换
- map()      某一列内容的映射

```python
df
```

```python
#将PWEEK的2周改为3周，将3周改为4周
df.replace([2,3],[3,4])
```

```python
df.replace({
    2:3,
    3:4,
    '世宏鹏':'世宏朋'
})
```

```python
#生成mapper映射的字典，可以使用字典推导式
mapper = {index:'p%s'%(str(index+1).rjust(3,'0')) for index in df.index}
#axis=0表示，前面mapper字典修改的是行索引，反之，为1则修改为列索引
df.rename(mapper,axis=0)
```

```python
df.rename({
    'PSOURCE':'CODE',
    'PDOC':'DOC',
},axis=1)
```

```python
#根据PWEEK的周数，生成一个新的列SPEED，映射的关系如下：
#2  -快 ，3-慢
df['SPEED']=df['PWEEK'].map({
    2:'快',
    3:'慢',
})
df
```

```python
#map(lambda函数)
df['SPEED']=df['PWEEK'].map(lambda week:'快'if week <= 2 else '慢')
df
```

假设张三李四的课表里有满分的情况，
老师认为是作弊，把所有满分的情况（包括150,300分）都记0分，如何实现？


练习：
新增两列，分别为张三、李四的成绩状态，如果分数低于90，则为"failed"，如果分数高于120，则为"excellent"，其他则为"pass"
【提示】使用函数作为map的参数

```python
data=np.random.randint(150,size=(4,3))
index=['张三','李四','王五','赵六']
columns=['Python','H5','MySQL']
df3=DataFrame(data,index,columns)
df3
```

```python
df3.loc['张三','Python'] =150
df3.loc['李四','MySQL'] =300
df3
```

```python
df3.replace({150:0,300:0})
```

练习： 新增两列，分别为张三、李四的成绩状态，如果分数低于90，则为"failed"，如果分数高于120，则为"excellent"，其他则为"pass" 【提示】使用函数作为map的参数

```python
df4 =df3.T
df4
```

```python
valid_score = lambda score:'failed' if score < 90 else 'excellent' if score > 120 else 'pass'
```

```python
df4['s1'] = df4['张三'].map(valid_score)
df4['s2'] = df4['李四'].map(valid_score)
df4
```

练习：
假设菜市场张大妈在卖菜，有以下属性：
菜品(item)：萝卜，白菜，辣椒，冬瓜
颜色(color)：白，青，红
重量(weight)
价格(price)
要求以属性作为列索引，新建一个df = DataFrame
对df进行聚合操作，求出颜色为白色的价格总和
对df进行聚合操作，求出萝卜的所有重量(包括白萝卜，胡萝卜，青萝卜）以及平均价格

```python
df7 = DataFrame({
    'item': np.random.choice(['萝卜','白菜','辣椒','冬瓜'],size=20),
    'color': np.random.choice(['白','青','红'],size=20),
    'weight(kg)': np.round(np.random.uniform(3,10,size=20),1),
    'price(RMB)': np.round(np.random.uniform(2.5,10.5,size=20),1)
})
df7
```

```python
#对df进行聚合操作，求出颜色为白色的价格总和
group7_color = df7.groupby('color')
group7_color
```

```python
group7_color['price(RMB)'].sum()
```

```python
#对df进行聚合操作，求出萝卜的所有重量
#(包括白萝卜，胡萝卜，青萝卜）以及平均价格
df7.groupby('item')['weight(kg)'].sum()
```

```python
df7.groupby('item')['price(RMB)'].mean()
```

```python
def apply_data(item):
    return Series({
        'weight(kg)_total':item['weight(kg)'].sum(),
        'price(RMB)_mean': item['price(RMB)'].mean()
    })
```

```python
df7[['weight(kg)','price(RMB)']]
```

```python
Series({
        'weight(kg)': df7['weight(kg)'].sum(),
        'price(RMB)': df7['price(RMB)'].mean()
    })
```

```python
df7.groupby('item')['weight(kg)','price(RMB)'].apply(apply_data)
```
