```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
```

读取数据

```python
aap1 = pd.read_csv('data/AAPL.csv')
aap1.head()
```

```python
aap1.dtypes
```

将'Date'这行数据转换为时间数据类型

```python
from datetime import datetime
```

```python
aap1['Date'] = aap1['Date'].map(lambda item:datetime.strptime(item,'%Y-%m-%d') )
aap1.head()
```

```python
aap1.dtypes
```

map()和transform()的区别:
    1.map()能用字典，transform()不能用字典
    2.map()只能用于单列(未分组)，transform()可以用于分组后的列值映射

```python
#使用pd.to_datetime()函数
pd.to_datetime(aap1['Date'])
```

```python
aap1['Date'] = pd.to_datetime(aap1['Date'])
aap1.dtypes
```

```python
aap1.set_index('Date')
```
